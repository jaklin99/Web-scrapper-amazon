```python
from bs4 import BeautifulSoup as bs4
from requests import get
from glob import glob
import pandas as pd
from datetime import datetime
from time import sleep
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

```python
amazon_urls = []
with open('input.txt', mode='r') as file:
    for line in file.readlines():
        amazon_urls.append(line.strip())
        print(line)
```

```python
amazon_urls
```

```python
soup_array=[]
for url in amazon_urls:
    # getting the reponse from the page using get method of requests module
    page = get(url, headers=HEADERS)

    # storing the content of the page in a variable
    html = page.content

    # creating BeautifulSoup object
    soup = bs4(html, "html.parser")
    soup_array.append(soup)
    print(soup_array)
```

```python
products = []

for i in soup_array:
    
    product_title = i.find('span', id = 'productTitle').text.strip()
    product_description = i.find('div', id = 'productDescription').text.strip()
    image = i.find('div', class_ = 'imgTagWrapper').img.get('src')
    price = i.find('span', id ='price_inside_buybox').text.strip()
   
    reviews_urls = []
    with open('ReviewsCount.txt', mode='r') as file:
        for line in file.readlines():
            reviews_urls.append(line.strip())
            print(line)
    reviews = []
    for url in reviews_urls:
        page = get(url, headers=HEADERS)
        html = page.content
        # creating BeautifulSoup object
        soup = bs4(html, "html.parser")
        reviews.append(soup)
    for r in reviews:
        review_ = r.find('div', class_ = 'a-row a-spacing-base a-size-base').text.split()
        review_count = review_[4] + ' ' + review_[5] + ' ' + review_[6]
        print(review_count)
        
        
    rating = i.find('span', class_ = 'a-size-medium a-color-base').text
    stock_status = i.find('div', id = 'availability').text.strip()
    delivery_check = i.find('table', id = 'tabular-buybox-container')
    #print(type(delivery_check))
    
    rows =  delivery_check.find_all('tr')
    shipper = rows[0].find_all('td')[1].text
    seller = rows[1].find_all('td')[1].text
    vars = product_title, product_description, image, price,  review_count, rating, stock_status
    
    if shipper:
        #print(shipper)
        vars = (*vars, shipper)
    else:
        print('nw')
    if seller:
        vars = (*vars, seller)
    else: 
        print('not working')
    
    attributes = i.find('table', id = 'productDetails_techSpec_section_2')
    
    if attributes:
        product_attributes = attributes.find_all('tr')
        attributes_ = []
        for row in product_attributes:
            attributes_.append(row.text.strip())
        if len(attributes_) > 0:
                #print(shipper)
            vars = (*vars, attributes_)
        else:
            print('not working')
            
    about_item = i.find('div', id = 'feature-bullets').text.strip()
    
    if about_item:
        vars = (*vars, about_item)
    else:
        print('not working')
        
    a_choice = i.find('span', class_ = 'a-size-small aok-float-left ac-badge-rectangle')

    def if_present(label):
        if label == a_choice:
            print('exists')
            return True
        else:
            print('no')
        return False
    
    present = if_present(a_choice)
    if present == True:
        choice_items = i.find_all('span')
        labels = []
        for c in choice_items:
            labels.append(c.text.strip())
        if len(labels) > 0:
            vars = (*vars, labels)
        else:
            print('not present')
        
    products.append(vars)

print(products[0])
      
```

```python
dataframe = pd.DataFrame(products)
dataframe.columns=['Product Title', 'Product Description', 'Thumbnail Image URL', 'Price', 'Reviews Count', 'Rating', 'Stock Status', 'Shipped by', 'Sold by', 'Product attribute', 'About item', 'Amazon choice label' ] 
dataframe 
```
